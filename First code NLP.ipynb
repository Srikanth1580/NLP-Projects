```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('learning')
```

```python
ps.stem('playing')
```

```python
ps.stem('marketing')
```

```python
ps.stem('maximimum')
```

```python
ps.stem('affectionately')
```

```python
words_to_stem = ['learn', 'learning', 'learned', 'plays', 'playing', 'played', 'marketing', 'maximimum', 'affectionately']
for word in words_to_stem:
    print(f"{word} --> {ps.stem(word)}")
```

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()

for word in words_to_stem:
    print(f"{word} --> {ls.stem(word)}")
```

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')

for word in words_to_stem:
    print(f"{word} --> {ss.stem(word)}")
```

```python
sent = "This is a sample sentence, showing off the stop words filtration."
sent_tokens = word_tokenize(sent)
sent_tokens
```

```python
for token in sent_tokens:
    print(nltk.pos_tag([token]))
```

```python
for token in sent_tokens:
    if token in stopwords.words('english'):
        print(token)
```

```python
from nltk import ne_chunk
```

```python
NE = "Apple is looking at buying U.K. startup for $1 billion"
```

```python
NE_tokens = word_tokenize(NE)
NE_tokens
```

```python
NE_Tags = nltk.pos_tag(NE_tokens)
NE_Tags
```

```python
NE_NER = nltk.ne_chunk(NE_Tags)
print(NE_NER)
```
